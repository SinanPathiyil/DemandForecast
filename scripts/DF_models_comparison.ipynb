```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# ---------------- Loop over Products ----------------
for product in products:
    print(f"\n========== Forecasting for {product} ==========")
    product_df = df[df['Product_Name']==product].copy()

    # Aggregate duplicates
    product_df = product_df.groupby('Date')['Total_Quantity'].sum().reset_index()
    product_df = product_df.set_index('Date').asfreq('W-MON', fill_value=0)
    y = product_df['Total_Quantity']

    # Train-test split
    train_size = int(len(y) * 0.8)
    y_train, y_test = y[:train_size], y[train_size:]

    results = []
    preds = {}  # to store predictions for plotting

    # ---------------- ARIMA ----------------
    try:
        arima_model = ARIMA(y_train, order=(1,1,1)).fit()
        pred_arima = arima_model.forecast(steps=len(y_test))
        results.append(('ARIMA', mean_absolute_error(y_test, pred_arima), np.sqrt(mean_squared_error(y_test, pred_arima))))
        preds['ARIMA'] = pred_arima
    except:
        results.append(('ARIMA', None, None))

    # ---------------- SARIMA ----------------
    try:
        sarima_model = SARIMAX(y_train, order=(1,1,1), seasonal_order=(1,1,1,52)).fit(disp=False)
        pred_sarima = sarima_model.forecast(steps=len(y_test))
        results.append(('SARIMA', mean_absolute_error(y_test, pred_sarima), np.sqrt(mean_squared_error(y_test, pred_sarima))))
        preds['SARIMA'] = pred_sarima
    except:
        results.append(('SARIMA', None, None))

    # ---------------- Prophet ----------------
    try:
        prophet_df = y_train.reset_index().rename(columns={'Date':'ds','Total_Quantity':'y'})
        prophet_model = Prophet(weekly_seasonality=True, daily_seasonality=False)
        prophet_model.fit(prophet_df)
        future = prophet_model.make_future_dataframe(periods=len(y_test), freq='W')
        forecast = prophet_model.predict(future)
        pred_prophet = forecast['yhat'].iloc[-len(y_test):].values
        results.append(('Prophet', mean_absolute_error(y_test, pred_prophet), np.sqrt(mean_squared_error(y_test, pred_prophet))))
        preds['Prophet'] = pred_prophet
    except:
        results.append(('Prophet', None, None))

    # ---------------- ETS ----------------
    try:
        ets_model = ExponentialSmoothing(y_train, trend='add', seasonal='add', seasonal_periods=52).fit()
        pred_ets = ets_model.forecast(len(y_test))
        results.append(('ETS', mean_absolute_error(y_test, pred_ets), np.sqrt(mean_squared_error(y_test, pred_ets))))
        preds['ETS'] = pred_ets
    except:
        results.append(('ETS', None, None))

    # ---------------- XGBoost ----------------
    try:
        lags = 3
        X_train = pd.DataFrame({f'lag_{i}': y_train.shift(i) for i in range(1,lags+1)}).dropna()
        y_train_xgb = y_train[lags:]
        X_test = pd.DataFrame({f'lag_{i}': y_test.shift(i) for i in range(1,lags+1)}).dropna()
        y_test_xgb = y_test[lags:]
        xgb_model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100)
        xgb_model.fit(X_train, y_train_xgb)
        pred_xgb = xgb_model.predict(X_test)
        results.append(('XGBoost', mean_absolute_error(y_test_xgb, pred_xgb), np.sqrt(mean_squared_error(y_test_xgb, pred_xgb))))
        preds['XGBoost'] = (pred_xgb, y_test_xgb)  # tuple for plotting
    except:
        results.append(('XGBoost', None, None))

    # ---------------- Croston ----------------
    try:
        pred_croston = croston(y_train.tolist())[-len(y_test):]
        results.append(('Croston', mean_absolute_error(y_test, pred_croston), np.sqrt(mean_squared_error(y_test, pred_croston))))
        preds['Croston'] = pred_croston
    except:
        results.append(('Croston', None, None))

    # ---------------- LSTM ----------------
    try:
        scaler = MinMaxScaler()
        y_scaled = scaler.fit_transform(y.values.reshape(-1,1))
        train_scaled = y_scaled[:train_size]
        test_scaled = y_scaled[train_size:]

        X_train_lstm, y_train_lstm = create_sequences(train_scaled, time_steps=3)
        X_test_lstm, y_test_lstm_seq = create_sequences(test_scaled, time_steps=3)

        model = Sequential()
        model.add(LSTM(50, activation='relu', input_shape=(X_train_lstm.shape[1],1)))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mse')
        model.fit(X_train_lstm, y_train_lstm, epochs=50, batch_size=1, verbose=0)

        y_pred_lstm_scaled = model.predict(X_test_lstm)
        y_pred_lstm = scaler.inverse_transform(y_pred_lstm_scaled)
        y_test_actual = y[train_size + 3:]  # align test
        results.append(('LSTM', mean_absolute_error(y_test_actual, y_pred_lstm), np.sqrt(mean_squared_error(y_test_actual, y_pred_lstm))))
        preds['LSTM'] = y_pred_lstm
    except:
        results.append(('LSTM', None, None))

    # ---------------- Display Results ----------------
    print("\nModel Comparison:")
    for name, mae, rmse in results:
        print(f"{name:10s} | MAE: {mae if mae is not None else 'NA'} | RMSE: {rmse if rmse is not None else 'NA'}")

    # ---------------- Forecast Plot ----------------
    plt.figure(figsize=(12,6))
    plt.plot(y_test.index, y_test, label='Actual', marker='o')
    if 'ARIMA' in preds: plt.plot(y_test.index, preds['ARIMA'], label='ARIMA')
    if 'SARIMA' in preds: plt.plot(y_test.index, preds['SARIMA'], label='SARIMA')
    if 'Prophet' in preds: plt.plot(y_test.index, preds['Prophet'], label='Prophet')
    if 'ETS' in preds: plt.plot(y_test.index, preds['ETS'], label='ETS')
    if 'XGBoost' in preds:
        plt.plot(preds['XGBoost'][1].index, preds['XGBoost'][0], label='XGBoost')
    if 'Croston' in preds: plt.plot(y_test.index, preds['Croston'], label='Croston')
    if 'LSTM' in preds: plt.plot(y_test_actual.index, preds['LSTM'], label='LSTM')
    plt.legend()
    plt.title(f"Forecast Comparison for {product}")
    plt.show()

    # ---------------- Bar Plot for Metrics ----------------
    model_names = [r[0] for r in results]
    maes = [r[1] if r[1] is not None else 0 for r in results]
    rmses = [r[2] if r[2] is not None else 0 for r in results]

    x = np.arange(len(model_names))
    width = 0.35
    plt.figure(figsize=(10,5))
    plt.bar(x - width/2, maes, width, label='MAE')
    plt.bar(x + width/2, rmses, width, label='RMSE')
    plt.xticks(x, model_names, rotation=45)
    plt.ylabel("Error")
    plt.title(f"Error Metrics for {product}")
    plt.legend()
    plt.show()
```
